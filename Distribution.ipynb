Question 4: What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

Answer: Standardization is a preprocessing technique used in statistics and data science to rescale data so that it has a mean of 0 and a standard deviation of 1. This results in a z-score, which represents how many standard deviations a specific data point is from the average of the dataset.

Importance in Data Analysis
Comparability: It allows you to compare data points from different scales or distributions (e.g., comparing a score on a math test out of 100 with a score on a physics test out of 50).

Outlier Detection: Since the mean is 0 and the standard deviation is 1, any z-score significantly far from 0 (typically >3 or <-3) can be easily identified as an outlier.

Algorithm Performance: Many machine learning algorithms (like K-Nearest Neighbors, SVM, and Neural Networks) perform better or converge faster when features are on the same scale, as they often rely on distance calculations.

How it Transforms a Dataset

Centering: Subtracting the mean
shifts the entire distribution so that the new center of the data is 0.

Scaling: Dividing by the standard deviation "squashes" or "stretches" the data so that the spread is consistent, resulting in a standard deviation of 1.

Preserving Shape: Importantly, standardization changes the scale and location of the data but does not change the shape of the original distribution (e.g., if it was skewed, it remains skewed).


Question 5: What is Kurtosis and their type?

Answer: In statistics, kurtosis is a measure used to describe the "tailedness" of a probability distribution. It quantifies how much of the data is concentrated in the tails relative to the peak, essentially indicating the presence of outliers or the heaviness of the tails compared to a normal distribution.

There are three primary types of kurtosis:
Mesokurtic: This describes a distribution that is identical to a normal distribution. Its tails are neither too thin nor too thick, and its "excess kurtosis" is zero (the raw kurtosis is exactly 3).

Leptokurtic (Positive Kurtosis): These distributions have heavy tails and a higher, sharper peak. This indicates a higher frequency of extreme outliers (thick tails) compared to a normal distribution.

Platykurtic (Negative Kurtosis): These distributions have thin tails and a flatter peak. The data is more spread out and less likely to produce extreme outliers, appearing "flatter" than a normal curve.

Question 6: Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

Answer: The uniform distribution is a good model for rolling a fair die because it describes a situation where every possible outcome has an equal probability of occurring.

Here is a breakdown of why this model fits:

Equal Likelihood: By definition, a "fair" die is not weighted or biased. This means that each of the six faces (1, 2, 3, 4, 5, and 6) has exactly the same chance of landing face up.

Constant Probability: Since there are 6 possible outcomes and each is equally likely, the probability for any single result x is constant:

P (X=x) = 1/6

Discrete Outcomes: This specific case is a Discrete Uniform Distribution because the outcomes are distinct, countable individual values rather than a continuous range.

Predictability of Symmetry: Because the physical properties of a fair die are symmetrical, the distribution of results over a large number of trials will "uniformly" flatten out, showing no preference for any specific number.

Question 8: How does log transformation help in stabilizing variance and making data more normally distributed?

Answer: A log transformation is a mathematical operation used in statistics to address data that violates assumptions like constant variance (homoscedasticity) and normality.

Stabilizing Variance
In many real-world datasets, the variance of a variable increases as its mean increases (heteroscedasticity).

Compression of Large Values: The logarithm function is non-linear; it compresses higher values much more than lower values. For example, the difference between 100 and 10 is
90, but their log-transformed values (base 10) are
1 and 2 ,a difference of only
1.

Uniform Spread: By pulling extreme values closer to the center, the overall spread (variance) becomes more uniform across the range of the data, helping it meet the requirements of models like linear regression.

Making Data More Normally Distributed

Log transformations are particularly effective for right-skewed (positively skewed) data, which has a long tail of high values.

Correcting Skewness: Because the log function "shrinks" the right tail of a distribution more than the left, it shifts the mass of the data toward the center.

Symmetry: This stretching of lower values and squeezing of higher values often results in a distribution that is more symmetrical and bell-shaped, resembling a normal distribution.
